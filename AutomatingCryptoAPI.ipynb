```python
def api_runner():
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
       'start':'1',
       'limit':'15',
       'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': '7b72412a-e769-4dd4-8dcd-353e0f3f6898',
    }

    session = Session()
    session.headers.update(headers)

    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      #print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)
   #appending the data in the loop
    df2 = pd.json_normalize(data['data'])
    df2['StampTime'] = pd.to_datetime('now')
    df_append = pd.DataFrame(df2)
    df = pd.concat([df,df_append])


    
```

```python
import os 
from time import time
from time import sleep

for i in range(333):
    api_runner()
    print("API Runner completed")
    sleep(30) #sleep for 1 minute
exit()
```

```python
#sorting data in the hourly and day format for further analysis
df3 = df.groupby('name', sort=False)[['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d']].mean()
df3
```

```python
#changing the columns to rows by stacking
df4=df3.stack()
df4
```

```python
#changing the type from series to dataframe which has been changed when we stacked the data earlier
df5= df4.to_frame(name='values')
df5
```
